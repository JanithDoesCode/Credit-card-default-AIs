```python
import pickle
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.tree import plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, confusion_matrix
)
```

```python
# Encoding for Age
age_label = df_data_v2['AGE'].unique()
print(age_label)
```

```python
age = df_data['AGE'].values
print(age)
```

```python
ages = []
for data in age:
    if data == 'Less than 30':
        ages.append(1)
    elif data == '31-45':
        ages.append(2)
    elif data == '46-65':
        ages.append(3)
    else:
        ages.append(4)
```

```python
df_data_v2['AGE'] = pd.DataFrame(ages)
df_data_v2.head()
```

```python
# Encoding for EDUCATION_STATUS
edu_status_label = df_data_v2['EDUCATION_STATUS'].unique()
print(edu_status_label)
```

```python
edu_status = df_data['EDUCATION_STATUS'].values
print(edu_status)
```

```python
upd_edu_status = []
for data in edu_status:
    if data == 'High School':
        upd_edu_status.append(1)
    elif data == 'Graduate':
        upd_edu_status.append(2)
    else:
        upd_edu_status.append(3)
```

```python
df_data_v2['EDUCATION_STATUS'] = pd.DataFrame(upd_edu_status)
df_data_v2.head()
```

```python
# Encoding for Balance_Limit_V1
balance_label = df_data_v2['Balance_Limit_V1'].unique()
print(balance_label)
```

```python
balance = df_data['Balance_Limit_V1'].values
print(balance)
```

```python
updated_balance = []
for value in balance:
    if value == '100K':
        updated_balance.append(100000)
    elif value == '200K':
        updated_balance.append(200000)
    elif value == '300K':
        updated_balance.append(300000)
    elif value == '400K':
        updated_balance.append(400000)
    elif value == '500K':
        updated_balance.append(500000)
    elif value == '1M':
        updated_balance.append(1000000)
    elif value == '1.5M':
        updated_balance.append(1500000)
    else:
        updated_balance.append(2500000)
```

```python
df_data_v2['Balance_Limit_V1'] = pd.DataFrame(updated_balance)
df_data_v2.head()
```

```python
df_data_v3 = df_data_v2.copy()
df_data_v3.head()
```

```python
# Drop unnecessary fields
df_data_v3.drop(['PAY_JULY', 'PAY_AUG', 'PAY_SEP', 'PAY_OCT', 'PAY_NOV'], axis=1, inplace=True)
df_data_v3.drop(['DUE_AMT_JULY', 'DUE_AMT_AUG', 'DUE_AMT_SEP', 'DUE_AMT_OCT', 'DUE_AMT_NOV'], axis=1, inplace=True)
df_data_v3.drop(['PAID_AMT_JULY', 'PAID_AMT_AUG', 'PAID_AMT_SEP', 'PAID_AMT_OCT', 'PAID_AMT_NOV'], axis=1, inplace=True)
df_data_v3.head()
```

```python
# Count of each unique category
print(df_data_v3['NEXT_MONTH_DEFAULT'].value_counts())
```

Split the dataset

```python
x = df_data_v3.drop(['NEXT_MONTH_DEFAULT'], axis=1)
y = df_data_v3['NEXT_MONTH_DEFAULT']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

Standardizing the independent variables

```python
standard = preprocessing.StandardScaler()
```

```python
x_train = standard.fit_transform(x_train)
x_test = standard.fit_transform(x_test)
```

```python
print(x_train)
```

Train the model

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'max_features': ['sqrt', 'log2']
}
```

```python
random_search = RandomizedSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_distributions=param_grid,
    n_iter=50,  # number of parameter settings sampled
    cv=5,
    n_jobs=-1,
    verbose=2,
    random_state=42
)
random_search.fit(x_train, y_train)
```

```python
print("Best parameters (RandomizedSearchCV):", random_search.best_params_)
rs_best_model = random_search.best_estimator_
```

```python
# Training accuracy
y_train_pred = rs_best_model.predict(x_train)

# Testing accuracy
y_test_pred = rs_best_model.predict(x_test)
```

Testing the model

```python
print("Training Accuracy:", accuracy_score(y_train, y_train_pred))

print("Testing Accuracy:", accuracy_score(y_test, y_test_pred))

print("Precision:", precision_score(y_test,
                                    y_test_pred,
                                    average="weighted"))

print('Recall:', recall_score(y_test,
                              y_test_pred,
                              average="weighted"))

print('F1 score:', f1_score(y_test, y_test_pred,
                            average="weighted"))

confusion_matrix = confusion_matrix(y_test, y_test_pred)
print('Confusion matrix:')
print(confusion_matrix)
```
